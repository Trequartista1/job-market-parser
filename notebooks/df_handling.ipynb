```python



def parse_skills(x):
    # если строка → пробуем преобразовать
    if isinstance(x, str):
        try:
            parsed = ast.literal_eval(x)
            return parsed if isinstance(parsed, list) else []
        except:
            return []

    # если уже список
    if isinstance(x, list):
        return x

    # всё остальное → пусто
    return []

```

```python
# 1. преобразуем в списки
res["skills"] = res["skills"].apply(parse_skills)

# 2. нормализуем (lower + trim)
res["skills"] = res["skills"].apply(
    lambda x: [s.strip().lower() for s in x if isinstance(s, str)]
)
```

```python

SKILL_MAP = {
    "python": ["python"],
    "sql": ["sql", "postgres", "mysql", "pl/sql"],
    "excel": ["excel", "google sheets", "sheets"],
    "power bi": ["power bi", "powerbi"],
    "tableau": ["tableau"],
    "pandas": ["pandas"],
    "numpy": ["numpy"],
    "statistics": ["statistics"],
    "machine learning": ["machine learning", "ml"],
    "etl": ["etl"],
    "api": ["api"],
    "git": ["git", "github", "gitlab"],
    "aws": ["aws"],
    "gcp": ["gcp"],
    "azure": ["azure"]
}

def process_skills(x):
    # 1. строка → список
    if isinstance(x, str):
        try:
            skills_list = ast.literal_eval(x)
        except:
            return []
    elif isinstance(x, list):
        skills_list = x
    else:
        return []

    # 2. нормализация + фильтрация
    result = set()

    for skill in skills_list:
        if not isinstance(skill, str):
            continue

        s = skill.lower().strip()

        for normalized, variants in SKILL_MAP.items():
            if any(v in s for v in variants):
                result.add(normalized)
                break

    return list(result)

# применяем
res["skills"] = res["skills"].apply(process_skills)

# удаляем пустые
res = res[res["skills"].apply(len) > 0]
```
